# 1. Inventory Price Calculator

```python
def calculate_inventory_summary(prices):
  total_price = 0
  average_price = 0
  for price in prices:
    total_price += price
  average_price = total_price/len(prices)
  return total_price,average_price
```

```python
total_price,average_price = calculate_inventory_summary([120,250,399,150])
print(f"({total_price},{average_price})")
```

# 2. Unique Visitor Tracker

```python
def count_unique_visitors(visitor_list):
  unique_visitors = len(set(visitor_list))
  return unique_visitors,sorted(set(visitor_list))
```

```python
unique_visitors,unique_IDs = count_unique_visitors([101,205,101,310,205,550])
print(f"{unique_visitors} ( The unique IDs are",end =" ")
for i in unique_IDs:
  print(i,end=" ")
print(")")
```

# 3. Product Category Counter

```python
def count_categories(products):
  categories = {}
  for product in products:
    if product in categories:
      categories[product]+=1
    else:
      categories[product]=1
  return categories
```

```python
categories = count_categories(['Laptop','Smartphone','Monitor','Laptop','Smartphone','Headphones','Laptop'])
print(categories)
```

# 4. Temperature Analytics

```python
def analyze_temperatures(temp_list):
  High_temp_list = []
  Normal_temp_list =[]
  for temp in temp_list:
    if temp > 38:
      High_temp_list.append(temp)
    else:
      Normal_temp_list.append(temp)
  return High_temp_list,Normal_temp_list
```

```python
High_temp_list,Normal_temp_list = analyze_temperatures([36.5,38.0,39.1,37.2,40.5,38.0,37.9])
print(f"High List (>38째C): {High_temp_list}")
print(f"Normal List (<38째C): {Normal_temp_list}")
print(f"Full Output Tuple: ({High_temp_list}, {Normal_temp_list})")
```

# 5. Menu Price Lookup

```python
def get_price(menu,item):
  item = item.capitalize()
  if item in menu:
    print("Item Found")
    return menu[item]
  else:
    return "Item Not Found"
```

```python
price = get_price({"Coffee": 3.50, "Tea": 2.75, "Muffin": 4.00, "Sandwich": 8.50},"Muffin")
print(price)
```

```python
price = get_price({"Coffee": 3.50, "Tea": 2.75, "Muffin": 4.00, "Sandwich": 8.50},"Donut")
print(price)
```

# 6. Student Grading System

```python
def calculate_grade(score):
  for name in score.keys():
    if score[name] > 100 or score[name] < 0:
      return "Invalid marks! Check the marks again!"
    elif score[name] >=80 and score[name] <= 100:
      score[name] = "A+"
    elif score[name] >=70 and score[name] <= 79:
      score[name] = "A"
    elif score[name] >=60 and score[name] <= 69:
      score[name] = "A-"
    elif score[name] < 40:
      score[name] = "F"
  return score
```

```python
grade = calculate_grade({"Alice": 85, "Bob": 78, "Charlie": 32, "David": 69})
print(grade)
```

# 7. Discount Calculator

```python
def apply_discounts(products):
  price_after_discount = []
  for product in products:
    name, price, discount = product
    price_after_discount.append(price - (price*(discount/100)))
  return price_after_discount
```

```python
price_after_discount = apply_discounts([("Keyboard", 80.00, 20), ("Mouse", 25.00, 10), ("Monitor", 300.00, 5)])
print(price_after_discount)
```

# 8. Hashtag Generator

```python
def generate_hashtags(sentence):
  hashtag_list = []
  words = sentence.split()
  for word in words:
    word = word.lower()
    hashtag_list.append("#" + word)
  return hashtag_list
```

```python
hashtag_list = generate_hashtags("AI Data Science")
print(hashtag_list)
```

# 10. Shopping Cart Bill Calculator

```python
def calculate_total(cart, prices):
  total_amount = 0
  for item in cart:
    item = item.capitalize()
    if item in prices:
      total_amount += prices[item]
    else:
      print(f"{item} is not in prices list.")
  return total_amount
```

```python
total_amount = calculate_total(["Milk", "Bread", "Milk", "Cheese", "Bread"], {"Milk": 3.00, "Bread": 2.50, "Eggs": 4.00, "Cheese": 5.00})
print(f"(Total Bill Amount): {total_amount}")
```

```python
total_amount = calculate_total(["Milk", "Bread", "Milk", "Mango", "Bread"], {"Milk": 3.00, "Bread": 2.50, "Eggs": 4.00, "Cheese": 5.00})
print(f"(Total Bill Amount): {total_amount}")
```
