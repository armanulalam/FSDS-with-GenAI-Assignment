## 1. Email Validator (String Filtering + Loops) 

```python
email =input("Enter your email address: ")
counter = 0

for i in email:
    if i == '@':
        counter += 1
        break

if counter == 1 and (email.endswith(".com") or email.endswith(".org")):
    print("Valid Email")
else:
    print("Invalid Email")
```

## 2. Count Word Frequency in Customer Reviews 

```python
review_text = input("Enter the review text: ").lower()
lst = [i.lower() for i in review_text.split()]
lst_new = []

for i in lst:
    if i not in lst_new:
        lst_new.append(i)

print("{",end = "")
cnt = 1
for i in lst_new:
    print(f" '{i}'",end = "")
    print(":",review_text.split().count(i),end = " ")
    
    if cnt < len(lst_new)-1:
        print(",",end = "") 
    else:
        print(end = "")

    cnt+=1
    
print(end = "}")
```

## 3. Password Strength Checker 

```python
password = input("Enter your password: ")
special_char_list = ['!','"', '#', '$', '%', '&',"'",'(',')','*', '+', ',', '-', '.', '/', ':', ';', '<', '=', '>','?','@','[']

counter1 = 0 
counter2 = 0 
counter3 = 0 
counter4 = 0

for i in password:
    if i >= 'A' and i <= 'Z':
        counter1 += 1
    elif i >= 'a' and i <= 'z':
        counter2 += 1
    elif i >= '0' and i <= '9':
        counter3 += 1
    elif i >= '0' and i <= '9':
        counter3 += 1
    elif i in special_char_list:
        counter4 += 1

if (counter1 != 0) and (counter2 != 0) and (counter3 != 0) and (counter4 != 0):
    print("Strong Password")
else:
    print("Weak Password")
```

## 4. Username Generator 

```python
import random

number = random.randint(1,100)
number = str(number)

name = input("Enter your full name: ")
lst = []

for i in range(len(name)):
    if i < 3:
        lst.append(name[i].lower())
    elif i > len(name)-4:
        lst.append(name[i].lower())

for i in number:
    lst.append(i)

for i in lst:
    print(i, end = '')
```

## 5. SMS Word Counter

```python
text = input("Enter the text here: ")

counter_words = 0
counter_characters = 0

for i in text.split():
    counter_words += 1

for i in text:
    counter_characters += 1

if counter_characters > 160:
    print("Exceed limit! SMS limit is 160 characters.")
else:
    print(f"Words = {counter_words}, Characters = {counter_characters}, Status = Within Limit")
```

## 6. Remove Duplicates from Contact List 

```python
names_list = []
duplicate_free_contact_list = []
n = int(input("How many numbers you want to add in the name list? "))

for i in range(n):
    name = input(f"Enter person {i+1} name: ")
    names_list.append(name)

for i in names_list:
    if i not in duplicate_free_contact_list:
        duplicate_free_contact_list.append(i)

print(duplicate_free_contact_list)
```

## 7. Detect Suspicious Words in Text (Moderation)

```python
suspicious_wordlist = ['spam', 'scam', 'fake']
text = input("Enter the text here: ")

counter = 0

for i in text.split():
    if i.lower() in suspicious_wordlist:
        counter += 1

if counter != 0:
    print("Warning: Suspicious content detected")    
else:
    print("The text doesn't contain any suspicious content.")
```

## 9. Count Hashtags and Mentions (Social Media Analytics) 

```python
post = input("Enter the post: ")

counter_hashtags = 0
counter_mentions = 0

for i in post:
    if i == "#":
        counter_hashtags += 1
    elif i == "@":
        counter_mentions += 1
print(f"Hashtags = {counter_hashtags}, Mentions = {counter_mentions}")
```

## 10. Word Reversal for Data Encryption 

```python
message = input("Enter your message: ")

for i in message.split():
    reverse = i[::-1]
    print(reverse, end = " ")
```
