# Assignment 1: Temperature Check 

```python
current_temperature = float(input("Enter the current temperature : "))

if current_temperature < 10:
    print("It's too cold!")

elif current_temperature >= 10 and current_temperature <= 25:
    print("It's cool outside" )

else:
    print("It's hot outside!")
```

# Assignment 2: ATM Withdrawal System 

```python
account_balance  = float(input("Enter the amount of your account balance : "))
withdrawal_amount = float(input("Enter the amount of your withdrawal : "))

if account_balance < withdrawal_amount:
    print("Insufficient Balance")

else:
    print("Transaction Successful")
    remaining_balance = account_balance-withdrawal_amount
    print("Your remaining balance is",remaining_balance)
```

# Assignment 3: Student Grade Calculator

```python
number = int(input("Enter the marks you obtained: "))

if number >= 90 and number <= 100:
    print("A+")

elif number >= 80 and number <= 89:
    print("A")

elif number >= 70 and number <= 79:
    print("B")

elif number >= 60 and number <= 69:
    print("C")

else:
    print("Fail")
```

# Assignment 4: Bus Ticket Discount 

```python
age = int(input("Enter your age: "))

if age < 5:
    print("Free")

elif age >= 5 and age <=18:
    print("50% Discount")

elif age >= 60:
    print("30% Discount")

else:
    print("Full Price")
```

# Assignment 5: Restaurant Bill with Discount

```python
bill = int(input("Enter the amount of the bill: "))

if bill > 1000:
    print("apply 10% discount")
    bill = bill - (bill * (10/100))

else:
    print("no discount")

print("You have to pay",bill)
```

# Assignment 7: Traffic Light System 

```python
color = input("""
Select a traffic color:
1.red
2.yellow
3.green
""")

if color == 'red':
    print("Stop")

elif color == 'yellow':
    print("Ready to go")

elif color == 'green':
    print("Go")

else:
    print('Invalid Color')
```

# Assignment 8: Leap Year Checker 

```python
year = int(input("Enter the year you want to check as leap year or not : "))

if year % 400 == 0:
    print("Leap Year")

elif year % 4 ==0 and year % 100 != 0:
    print("Leap Year")

else:
    print("Not Leap Year")
```

# Assignment 10: Electricity Bill Calculator 

```python
unit_consumed = float(input("Enter the number of units consumed electricity : "))

if unit_consumed <= 100:
    bill_calculation = 5 * unit_consumed

elif unit_consumed >= 101 and unit_consumed <=200:
    bill_calculation = 7 * unit_consumed

else:
    bill_calculation = 10 * unit_consumed

print("Total payable amount is",bill_calculation, "BDT") 
```
